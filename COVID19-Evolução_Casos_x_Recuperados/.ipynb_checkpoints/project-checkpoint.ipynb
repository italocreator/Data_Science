## COVID19 - Evolução no numero de casos confirmados e recuperados 

```python
import requests as r
import datetime as dt
import csv
from PIL import Image
from IPython.display import display
```

```python
url = 'https://api.covid19api.com/dayone/country/brazil'
```

```python
resp = r.get(url)
```

```python
resp.status_code
```

```python
raw_data = resp.json()
```

```python
raw_data[0]
```

```python
final_data = []
for obs in raw_data:
    final_data.append([obs['Confirmed'], obs['Deaths'], obs['Recovered'], obs['Active'], obs['Date']])
```

```python
final_data[:10]
```

```python
final_data.insert(0, ['Confirmados', 'Obitos', 'Recuperados', 'Ativos', 'Data'])
```

```python
final_data[:5]
```

```python
CONFIRMADOS = 0
OBITOS = 1
RECUPERADOS = 2
ATIVOS = 3
DATA = 4
```

```python
# Preparando col datas
for i in range(1, len(final_data)):
    final_data[i][DATA] = final_data[i][DATA][:10]
final_data[:5]
```

```python
with open('brasil-covid.csv', 'w') as file:
    writer = csv.writer(file)
    writer.writerows(final_data)
```

```python
for i in range(1, len(final_data)):
    final_data[i][DATA] = dt.datetime.strptime(final_data[i][DATA], '%Y-%m-%d')
    
final_data[:10]
```

### QuickChart

```python
def get_datasets(y, labels):
    if type(y[0]) == list:
        datasets = []
        for i in range(len(y)):
            datasets.append({
                'label': labels[i],
                'data': y[i]
            })
        return datasets
    else:
        return [
            {
                'label': labels[0],
                'data': y
            }
        ]
```

```python
def set_title(title=''):
    if title != '':
        display = 'true'
    else:
        display = 'false'
    return {
        'title': title,
        'display': display
    }
```

```python
def create_chart(x, y, labels, kind='bar', title=''):
    
    datasets = get_datasets(y, labels)
    options = set_title(title)
    
    chart = {
        'type': kind,
        'data': {
            'labels': x,
            'datasets': datasets
        },
        'options': options
    }
    return chart
```

```python
def get_api_chart(chart):
    url_base = 'https://quickchart.io/chart'
    resp = r.get(f'{url_base}?c={str(chart)}')
    return resp.content
```

```python
def save_image(path, content):
    with open(path, 'wb') as image:
        image.write(content)
```

```python
def display_image(path):
    img_pil = Image.open(path)
    display(img_pil)
```

```python
#evolução a cada 10 dias
y_data_1 = []
for obs in final_data[1::10]:
    y_data_1.append(obs[CONFIRMADOS])
    if obs[CONFIRMADOS] >= 20034407:
        break
    
y_data_2 = []
for obs in final_data[1::10]:
    y_data_2.append(obs[RECUPERADOS])
    if obs[CONFIRMADOS] >= 20034407:
        break
    
labels = ['Confirmados', 'Recuperados']

x = []
for obs in final_data[1::10]:
    x.append(obs[DATA].strftime('%d/%m/%Y'))
    if obs[CONFIRMADOS] >= 20034407:
        break
    
chart = create_chart(x, [y_data_1, y_data_2], labels, title = 'Grafico confirmados x recuperados')
chart_content = get_api_chart(chart)
save_image('graph_confirmados_recuperados.png', chart_content)
display_image('graph_confirmados_recuperados.png')
```
