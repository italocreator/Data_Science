## Case Educação

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('Students.csv')
df.head()
```

```python
df.count()
```

```python
df.isnull().sum()
```

## Análise por Gênero

```python
df['gender'].value_counts()
```

```python
df_genero = df['gender'].value_counts(normalize=True)*100

plt.figure( figsize=(16,5) )
plt.title('Gênero', weight="bold", size=16)
plt.pie(
    df_genero,
    labels= df['gender'].unique(),
    shadow = True,
    startangle = 90,
    autopct = '%1.1f%%'
    )
plt.show()
```

## Resultado em provas por Gênero

### Matemática

```python
plt.figure( figsize=(5,7) )
ax = sns.boxplot(data = df, x = 'gender', y ='math score', linewidth=2 )

plt.title('Notas de Matemática por Gênero',dict(color ='black'), weight="bold", size=14)
plt.ylabel('Nota de Matemática',dict(color ='black'), size=12)
plt.xlabel('Gênero',dict(color ='black'), size=12)

ax.plot()
```

### Leitura

```python
plt.figure( figsize=(5,7) )
ax = sns.boxplot(data = df, x = 'gender', y ='reading score', linewidth=2, hue = "gender" )

plt.title('Notas de Leitura por Gênero',dict(color ='black'), weight="bold", size=14)
plt.ylabel('Nota de Leitura',dict(color ='black'), size=12)
plt.xlabel('Gênero',dict(color ='black'), size=12)

ax.plot()
```

### Escrita

```python
plt.figure( figsize=(5,7) )
ax = sns.boxplot(data = df, x = 'gender', y ='writing score', linewidth=2, hue = "gender")

plt.title('Notas de Escrita por Gênero',dict(color ='black'), weight="bold", size=14)
plt.ylabel('Nota de Escrita',dict(color ='black'), size=12)
plt.xlabel('Gênero',dict(color ='black'), size=12)

ax.plot()
```

### Alunos por etnia

```python
df_etnia = df['race/ethnicity'].value_counts(normalize=True)*100

plt.figure( figsize=(16,5) )
plt.title('Grupos de alunos por etnia', weight="bold", size=16)
plt.pie(
    df_etnia,
    labels= df['race/ethnicity'].unique(),
    shadow = True,
    startangle = 90,
    autopct = '%1.1f%%'
    )
plt.show()
```

### Distribuição de alunos por nível de educação dos pais

```python
dist_nivel_educ = df['parental level of education'].value_counts(normalize=True)*100

plt.figure( figsize=(16,5) )
plt.title('Nível de educação', weight="bold", size=16)
plt.pie(
    dist_nivel_educ,
    labels= df['parental level of education'].unique(),
    shadow = True,
    startangle = 90,
    autopct = '%1.1f%%'
    )
plt.show()
```

```python
df.groupby(by= ['parental level of education']).describe()['math score'].reset_index()
```

```python
df.groupby(by= ['parental level of education']).describe()['reading score'].reset_index()
```

```python
df.groupby(by= ['parental level of education']).describe()['writing score'].reset_index()
```

Aqui é possível verificar que independente da matéria testada, a média das notas dos estudantes tende ser maior quando os pais tem algum tipo de graduação.
